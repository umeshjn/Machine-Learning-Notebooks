Named-entity recognition is a subtask of information extraction that seeks to locate and classify named entity mentioned in unstructured text into pre-defined categories such as person names, organizations, locations, medical codes, time expressions, quantities, monetary values, percentages, etc.

NER is a very powerful text mining technique which is used for extraction of the real insights hidden in the text. 

```python
import spacy
from spacy import displacy

text = "Cases of the virus, a type of coronavirus, have spread in China, and have also been reported in nearby countries in Asia and in the United States and France."

nlp = spacy.load("en_core_web_md")
doc = nlp(text)
displacy.render(doc, style="ent", jupyter=True)
```

Above glimpse is the sample of what is possible using spacy. Data visualization is an amazing tool for showing the insights from the dataset but ability to visualize the entity recognition in text is never seen before. This is so powerful and easy to use tool in the hands of the data science engineers.

```python
### What are these entity labels
spacy.explain('GPE')
```

```python
spacy.explain('LOC')
```

#### What entities were used by President Trump in his State of the Union address of 2018?

```python
entities = []

for ent in doc.ents:
    entities.append(ent.label_)
    
print(entities[:10])
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (20,6)

keys, counts = np.unique(entities, return_counts=True)
```

```python
plt.bar(keys, counts)
plt.show()
```

Mr President has used lot GPE, ORG and Person entities. Let us get more insight into these entities.

```python
gpes = []

for ent in doc.ents:
    if ent.label_ == 'GPE':
        gpes.append(ent.text)

keys, counts = np.unique(gpes, return_counts=True)

plt.bar(keys, counts)
plt.xticks(rotation=90)
plt.show()
```

```python
persons = []

for ent in doc.ents:
    if ent.label_ == 'PERSON':
        persons.append(ent.text)

keys, counts = np.unique(persons, return_counts=True)

plt.bar(keys, counts)
plt.xticks(rotation=90)
plt.show()
```

```python
orgs = []

for ent in doc.ents:
    if ent.label_ == 'ORG':
        orgs.append(ent.text)

keys, counts = np.unique(orgs, return_counts=True)

plt.bar(keys, counts)
plt.xticks(rotation=90)
plt.show()
```
